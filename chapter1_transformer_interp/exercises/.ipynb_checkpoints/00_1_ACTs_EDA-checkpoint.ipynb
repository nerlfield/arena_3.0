```python
input_folder = './coffe_to_apple_images'
out_file = './coffe_to_apple.mp4'

N_FRAMES = 300  # Adjusted the repeated variable declaration

from PIL import Image, ImageDraw, ImageFont
import numpy as np
from moviepy.editor import ImageSequenceClip
from pathlib import Path; Path.ls = lambda x: list(x.iterdir())

# Sorting the files and selecting the number of frames needed
files = sorted(Path(input_folder).ls())
if N_FRAMES:
    files = files[:N_FRAMES]
print('len(files)', len(files))

# Load and process each image
frames = []
for i, f in enumerate(files):
    if f.suffix == '.png':
        img = Image.open(f)
        draw = ImageDraw.Draw(img)
        # You can customize the font size and type here
        font = ImageFont.truetype("arial.ttf", 20)  # Ensure the font file is available
        text = f"Attn Head {i}"
        draw.text((10, 10), text, font=font, fill="white")  # Adjust position and color as needed
        frames.append(np.array(img))

print(frames[0].shape)

fps = 25
out_video_path = out_file

# Create the video clip
clip = ImageSequenceClip(frames, fps=fps)
clip.write_videofile(out_video_path, fps=fps)
```

```python
import pickle

def write_pickle(filename, obj):
    with open(filename, 'wb') as handle:
        pickle.dump(obj, handle)

def read_pickle(filename):
    with open(filename, 'rb') as handle:
        return pickle.load(handle)
```

```python
import torch
```

```python
layer_head_diffs = {}

for layer in range(24):
    for head in range(24):
        diffs_list = []
        for k1 in list(MEGA_HIDDEN_STATES.keys()):
            for k2 in list(MEGA_HIDDEN_STATES.keys()):
                if k1 == k2: continue
                    
                diff_v = F.mse_loss(MEGA_HIDDEN_STATES[k1][(layer,head)], MEGA_HIDDEN_STATES[k2][(layer,head)]).item()
                diffs_list.append(diff_v)
                
        layer_head_diffs[(layer, head)] = torch.tensor(diffs_list).mean().item()
```

```python
write_pickle("layer_head_diffs.pickle", list(layer_head_diffs.items()))
```

```python
sorted(layer_head_diffs.items(), key=lambda x: x[1]).
```

```python
dick = {}

for (k, v), _ in sorted(layer_head_diffs.items(), key=lambda x: x[1])[:10]:
    dick[k] = dick.get(k, [])
    dick[k].append(v)
```
